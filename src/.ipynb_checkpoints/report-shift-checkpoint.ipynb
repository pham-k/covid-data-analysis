```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/prepared/shift.csv')
```

```python
df.date_report = pd.to_datetime(df.date_report, format='%d/%m/%Y')
df.date_positive = pd.to_datetime(df.date_positive, errors='coerce',format='%d/%m/%Y')
```

```python
inc = (df[['date_report']]
       .groupby(['date_report'])
       .apply(lambda x: len(x))
       .to_frame(name='case'))
        
inc['case_rm'] = inc.rolling(7).mean() 

inc_addr_dist = (df[['date_report', 'addr_dist']]
                 .groupby(['date_report', 'addr_dist'])
                 .apply(lambda x: len(x))
                 .to_frame(name='case')
                 .unstack(fill_value=0)
                 .asfreq('D', fill_value=0)
                 .stack()
                 .sort_index(level=0)
                 .reset_index(1)
                )


inc_addr_dist['case_rm'] = inc_addr_dist.groupby('addr_dist')['case'].transform(lambda x: x.rolling(7).mean())
inc_addr_dist = inc_addr_dist.reset_index(0)
```

```python
# inc_addr_dist[inc_addr_dist.addr_dist == 'QCCH']
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(inc.index, inc.case, label='Số ca mới theo ngày')
ax.plot(inc.index, inc.case_rm, label='Biến động trung bình 7 ngày')
ax.set_ylabel('Số ca mới')
ax.set_xlabel('Ngày')
ax.tick_params(axis='x', labelrotation=90)
ax.set_title('Số ca mới theo ngày và biến động trung bình 7 ngày')
ax.legend()

# inc = inc.reset_index()
# baz = pd.melt(inc, ['date_report'])
# baz

```

```python
g = sns.FacetGrid(
        inc_addr_dist.query('addr_dist in ["QA01", "QA02", "QA03", "QA04", "QA05", "QA06"]'),
        col="addr_dist", 
        col_wrap=3,
        height=3,
        aspect=3)
g.map(sns.lineplot, 'date_report', "case")
g.map(sns.lineplot, 'date_report', "case_rm", color='orange')
g.set_xticklabels(rotation=90)
g.set_axis_labels("Ngày", "Số ca mới")

# g = sns.FacetGrid(
#         foo,
#         col="addr_dist", 
#         col_wrap=3,
#         height=2,
#         aspect=3)
# g.map(sns.lineplot, 'date_report', 'value', 'variable')
# g.set_xticklabels(rotation=90)
# g.set_axis_labels("Ngày", "Số ca mới")
```
